```python
import os

import seaborn as sns

from skin_disease_recognition.config import RAW_DATA_DIR
```

```python
data_dir = RAW_DATA_DIR / 'SkinDisease' / 'train'
```

```python
classes = os.listdir(data_dir)
```

```python
counts = [len(os.listdir(data_dir / c)) for c in classes]
```

```python
for cls, count in zip(classes, counts, strict=True):
    print(cls, count)
```

```python
ax = sns.barplot(x=classes, y=counts)
ax.tick_params(labelrotation=60)
```
